```python
import numpy as np

temps = [32, 28, 35, 41, 29, 33, 37, 30, 26, 38,
         34, 31, 27, 36, 33, 29, 42, 31, 35, 28,
         30, 37, 33, 29, 34, 31, 36, 28, 32, 35]
```

```python
moyenne = np.mean(temps)
ecart_type = np.std(temps, ddof=1)  # ddof=1 car c'est un échantillon
n = len(temps)

print(f"Nombre de livraisons : n = {n}")
print(f"Moyenne : {moyenne:.2f} minutes")
print(f"Écart-type : {ecart_type:.2f} minutes")
```

```python
erreur_standard = ecart_type / np.sqrt(n)

print(f"Erreur standard : SE = {ecart_type:.2f} / √{n} = {erreur_standard:.2f} minutes")
```

```python
z = 1.96

borne_basse = moyenne - z * erreur_standard
borne_haute = moyenne + z * erreur_standard

print(f"IC à 95% = {moyenne:.2f} ± {z} × {erreur_standard:.2f}")
print(f"IC à 95% = {moyenne:.2f} ± {z * erreur_standard:.2f}")
print(f"IC à 95% = [{borne_basse:.2f}, {borne_haute:.2f}]")
```

```python
promesse = 35

print(f"IC à 95% = [{borne_basse:.2f}, {borne_haute:.2f}]")
print(f"Promesse = {promesse} minutes")
print()

if borne_haute < promesse:
    print(f"La borne haute ({borne_haute:.2f}) est SOUS les {promesse} min")
    print("→ OUI, la promesse est réaliste.")
    print("  Le temps moyen est significativement inférieur à 35 minutes.")
else:
    print(f"La borne haute ({borne_haute:.2f}) dépasse les {promesse} min")
    print("→ NON, la promesse n'est pas réaliste.")
```

```python
niveaux = [
    (0.90, 1.645),
    (0.95, 1.960),
    (0.99, 2.576)
]
for conf, z_val in niveaux:
    borne_basse = moyenne - z_val * erreur_standard
    borne_haute = moyenne + z_val * erreur_standard
    print(f"IC à {conf*100}% = {moyenne:.2f} ± {z_val} × {erreur_standard:.2f}")
    print(f"IC à {conf*100}% = {moyenne:.2f} ± {z_val * erreur_standard:.2f}")
    print(f"IC à {conf*100}% = [{borne_basse:.2f}, {borne_haute:.2f}]")
```
