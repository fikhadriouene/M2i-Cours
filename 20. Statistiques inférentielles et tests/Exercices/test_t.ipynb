```python
import numpy as np
from scipy import stats
from scipy.stats import shapiro, levene, mannwhitneyu

poids = [248, 253, 246, 251, 249, 255, 247, 252, 250, 248,
         254, 246, 251, 249, 253]

n = len(poids)
moyenne = np.mean(poids)
ecart_type = np.std(poids, ddof=1)

print(f"n = {n}")
print(f"Moyenne = {moyenne:.2f}g")
print(f"Écart-type = {ecart_type:.2f}g")
```

```python
stat, p_norm = shapiro(poids)
print(f"Shapiro-Wilk : p = {p_norm:.4f}")
```

```python
t_stat, p_value = stats.ttest_1samp(poids, popmean=250)

print(f"Statistique t = {t_stat:.3f}")
print(f"p-value = {p_value:.4f}")
```

```python
classique = [7, 6, 8, 5, 7, 6, 7, 8, 6, 5, 7, 6]
nouvelle  = [8, 9, 7, 8, 9, 8, 7, 9, 8, 9, 8, 7]

print(f"Recette classique : moyenne = {np.mean(classique):.2f}, écart-type = {np.std(classique, ddof=1):.2f}")
print(f"Nouvelle recette  : moyenne = {np.mean(nouvelle):.2f}, écart-type = {np.std(nouvelle, ddof=1):.2f}")
print(f"Différence : {np.mean(nouvelle) - np.mean(classique):.2f} points")
```

```python
stat_c, p_c = shapiro(classique)
stat_n, p_n = shapiro(nouvelle)

print(f"Shapiro classique : p = {p_c:.4f}")
print(f"Shapiro nouvelle  : p = {p_n:.4f}")
```

```python
stat_lev, p_lev = levene(classique, nouvelle)
print(f"Levene : p = {p_lev:.4f}")
```

```python
t_stat, p_value = stats.mannwhitneyu(classique, nouvelle)

print(f"Statistique t = {t_stat:.3f}")
print(f"p-value = {p_value:.4f}")
```

```python
n1, n2 = len(classique), len(nouvelle)
var1, var2 = np.var(classique, ddof=1), np.var(nouvelle, ddof=1)
pooled_std = np.sqrt(((n1-1)*var1 + (n2-1)*var2) / (n1+n2-2))

d = (np.mean(nouvelle) - np.mean(classique)) / pooled_std

print(f"Cohen's d = {d:.3f}")
```

```python
avant = [25, 28, 22, 30, 26, 24, 27, 23, 29, 25]
apres = [30, 33, 28, 34, 31, 29, 32, 28, 35, 30]

differences = np.array(apres) - np.array(avant)

print("Barista par barista :")
print(f"{'Barista':>8s} {'Avant':>6s} {'Après':>6s} {'Diff':>6s}")
print("-" * 30)
for i in range(len(avant)):
    print(f"{'B' + str(i+1):>8s} {avant[i]:6d} {apres[i]:6d} {differences[i]:+6d}")
```

```python
print(f"Différences : {list(differences)}")
print(f"Moyenne des différences : {np.mean(differences):.2f} cafés/heure")
print(f"Écart-type des différences : {np.std(differences, ddof=1):.2f}")
```

```python
stat, p_norm = shapiro(differences)
print(f"Shapiro-Wilk sur les différences : p = {p_norm:.4f}")
```

```python
t_stat, p_value = stats.mannwhitneyu(avant, apres)

print(f"Statistique t = {t_stat:.3f}")
print(f"p-value = {p_value:.4f}")
```
